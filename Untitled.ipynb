```python
#Load In Packages

import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OrdinalEncoder, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn import metrics
from sklearn .metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn .metrics import accuracy_score
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')

pd.set_option("display.max.columns", None)
```

```python
#Load In Data
df = pd.read_csv("../train.csv",sep = "," , encoding = "utf-8")
df_test = pd.read_csv("../test.csv",sep = "," , encoding = "utf-8")

def Month_Converter(x):
    if pd.notnull(x):
        num1 = int(x.split(' ')[0])
        num2 = int(x.split(' ')[3])
      
        return num1*12+num2
    else:
        return x
```

```python
#Preprocessing
df = df.applymap(lambda x: x if x is np.NaN or not isinstance(x, str) else str(x).strip('_ ,"')).replace(['', 'nan', '!@9#%8', '#F%$D@*&8'], np.NaN)
df['ID'] = df.ID.apply(lambda x: int(x, 16))
df['Customer_ID'] = df.Customer_ID.apply(lambda x: int(x[4:], 16))
df['Age'] = df.Age.astype(int)
df['SSN'] = df.SSN.apply(lambda x: x if x is np.NaN else float(str(x).replace('-', ''))).astype(float)
df['Annual_Income'] = df.Annual_Income.astype(float)
df['Num_of_Loan'] = df.Num_of_Loan.astype(int) 
df['Num_of_Delayed_Payment'] = df.Num_of_Delayed_Payment.astype(float)
df['Changed_Credit_Limit'] = df.Changed_Credit_Limit.astype(float)
df['Outstanding_Debt'] = df.Outstanding_Debt.astype(float)
df['Amount_invested_monthly'] = df.Amount_invested_monthly.astype(float)
df['Monthly_Balance'] = df.Monthly_Balance.astype(float)
df['Credit_History_Age'] = df.Credit_History_Age.apply(lambda x: Month_Converter(x)).astype(float)
df = df.drop(['Name','SSN','ID','Customer_ID'], axis=1)

month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}

df['Month'] = df['Month'].map(month_map)

le = LabelEncoder()
df['Occupation'] = le.fit_transform(df['Occupation'])

df = pd.get_dummies(df, columns=['Type_of_Loan', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour', 'Credit_Score'], drop_first=True)

print(df.dtypes)
```

```python
#Exploratory Data Analysis

import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].hist(df['Age'], bins=30, edgecolor='k')
axs[0, 0].set_title('Histogram of Age')

sns.boxplot(x=df['Annual_Income'], ax=axs[0, 1])
axs[0, 1].set_title('Box Plot of Annual Income')

axs[1, 0].scatter(df['Annual_Income'], df['Outstanding_Debt'])
axs[1, 0].set_title('Scatter Plot of Annual Income vs. Outstanding Debt')
axs[1, 0].set_xlabel('Annual Income')
axs[1, 0].set_ylabel('Outstanding Debt')

plt.tight_layout()
plt.show()
```

```python
#Correlation Matrix
selected_features = ['Age', 'Annual_Income', 'Num_of_Loan', 'Outstanding_Debt', 'Amount_invested_monthly', 'Monthly_Balance', 'Credit_History_Age']

corr_matrix = df[selected_features].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Feature Selection With Random Forest
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

df_sampled = df.sample(frac=0.01, random_state=42)

target_column = 'Annual_Income'
labels_sampled = df_sampled[target_column]
features_sampled = df_sampled.drop([target_column], axis=1)

categorical_columns = features_sampled.select_dtypes(include=['object']).columns
features_sampled[categorical_columns] = features_sampled[categorical_columns].fillna('missing')

features_encoded = pd.get_dummies(features_sampled, drop_first=True)

imputer = SimpleImputer(strategy='mean')
features_imputed = imputer.fit_transform(features_encoded)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(features_imputed, labels_sampled)

feature_importances = model.feature_importances_

feature_importance_df = pd.DataFrame({
    'Feature': features_encoded.columns,
    'Importance': feature_importances
})

feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(14, 10))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='skyblue')
plt.xlabel('Feature Importance', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.title('Feature Importances from Random Forest', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

important_features = feature_importance_df.head(3)['Feature'].tolist()

print("Selected important features:", important_features)
```

```python
#KNN

df_sampled = df.sample(frac=0.005, random_state=42)
labels_sampled = df_sampled['Annual_Income']
features_sampled = df_sampled.drop(['Annual_Income'], axis=1)

imputer = SimpleImputer(strategy='mean')
features_imputed = imputer.fit_transform(features_sampled)


target_column = 'Annual_Income'
features = df.drop([target_column], axis=1)
target = df[target_column]

features_imputed = imputer.fit_transform(features)

X_train, X_test, y_train, y_test = train_test_split(features_imputed, target, test_size=0.2, random_state=42)

print("Training Features Shape:", X_train.shape)
print("Testing Features Shape:", X_test.shape)
print("Training Labels Shape:", y_train.shape)
print("Testing Labels Shape:", y_test.shape)

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

knn_reg = KNeighborsRegressor(n_neighbors=5)
knn_reg.fit(X_train, y_train)

y_pred = knn_reg.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

k_values = range(1, 5)
mse_values = []

for k in k_values:
    knn_reg = KNeighborsRegressor(n_neighbors=k)
    knn_reg.fit(X_train, y_train)
    y_pred = knn_reg.predict(X_test)
    mse_values.append(mean_squared_error(y_test, y_pred))

plt.figure(figsize=(12, 6))
plt.plot(k_values, mse_values, marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Mean Squared Error')
plt.title('KNN Regressor Varying number of neighbors')
plt.show()
```

```python
#SVM
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

df_sampled = df.sample(frac=0.01, random_state=42)

target_column = 'Annual_Income'
labels_sampled = df_sampled[target_column]
features_sampled = df_sampled.drop([target_column], axis=1)

imputer = SimpleImputer(strategy='mean')
features_imputed = imputer.fit_transform(features_sampled)

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X_train, X_test, y_train, y_test = train_test_split(features_scaled, labels_sampled, test_size=0.2, random_state=42)

svm_regressor = SVR(kernel='rbf', C=100, gamma=0.1)  # You can tune C and gamma
svm_regressor.fit(X_train, y_train)

y_pred = svm_regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_pred, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted Annual Income')
plt.show()
```
